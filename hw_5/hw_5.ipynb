# Оптимизация типов pandas


```python
#данные
df1_pd= pd.read_csv('Housing.csv')
df1_pd
```

```python
#поиск оптимального типа данных, исходный размер 55,5KB
print(df1_pd.info())

#int64 можно перевести в int8/int16/int32, требующие меньше памяти
#для столбцов bedrooms,	bathrooms, stories, parking	- int8, числа в этих столбцах гарантированно не превышают 127

print('\nmax in area:',df1_pd['area'].max()) # 16200 < 32767 => можем применить int16
print('max in price',df1_pd['price'].max()) # 32767 < 13300000 < 2147483647 => можем применить int32

#для столбцов
def changeint_8(df):
  return df.astype(np.int8)

def changeint_16(df):
  return df.astype(np.int16)

def changeint_32(df):
  return df.astype(np.int32)

#object можно перевести в категориальный тип, который так же не занимает большой объем
def catchange(df):
  return df.astype("category")
```

```python
#переход к оптимальному типу данных полученный размер 10KB
df1_1_pd = df1_pd

df1_1_pd['price']= changeint_32(df1_1_pd['price'])
df1_1_pd['area']= changeint_16(df1_1_pd['area'])

df1_1_pd['bedrooms']= changeint_8(df1_1_pd['bedrooms'])
df1_1_pd['bathrooms']= changeint_8(df1_1_pd['bathrooms'])
df1_1_pd['stories']= changeint_8(df1_1_pd['stories'])
df1_1_pd['parking']= changeint_8(df1_1_pd['parking'])

df1_1_pd['mainroad']= catchange(df1_1_pd['mainroad'])
df1_1_pd['guestroom']= catchange(df1_1_pd['guestroom'])
df1_1_pd['basement']= catchange(df1_1_pd['basement'])
df1_1_pd['hotwaterheating']= catchange(df1_1_pd['hotwaterheating'])
df1_1_pd['airconditioning']= catchange(df1_1_pd['airconditioning'])
df1_1_pd['prefarea']= catchange(df1_1_pd['prefarea'])
df1_1_pd['furnishingstatus']= catchange(df1_1_pd['furnishingstatus'])

print(df1_1_pd.info())
```

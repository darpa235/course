#Работа с API

```python
import requests
import pandas as pd
```

```python
base_url = 'https://www.anapioficeandfire.com/api'

def ice_and_fire_lists(x, p):
  x_list= requests.get(
      f'{base_url}{x}',
  params=p).json()
  return(x_list)
```

```python
#данные о книгах
books_list=ice_and_fire_lists('/books','')

df_books = pd.DataFrame(books_list)
df_books
```

```python
#данные о домах
houses_list=ice_and_fire_lists('/houses','')

df_houses = pd.DataFrame(houses_list)
df_houses
```

```python
#данные о домах с титулами
houses_wt_list=ice_and_fire_lists('/houses', {'hasTitles': True})

df_houses_wt = pd.DataFrame(houses_wt_list)
df_houses_wt
```

```python
#данные о домах с девизами
houses_ww_list=ice_and_fire_lists('/houses', {'hasWords': True})

df_houses_ww = pd.DataFrame(houses_ww_list)
df_houses_ww
```

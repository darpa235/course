```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
```

```python
#формат ссылки
base_url='https://lifehacker.ru/'
response = requests.get('https://lifehacker.ru/topics/technology/')
soup = BeautifulSoup(response.text, 'lxml')
```

```python
#уникальные классы/идентификаторы блока + ссылки
raw_items = soup.find_all('a', class_='lh-small-article-card__link')

links = [item.get('href') for item in raw_items]
links = links[0:10]
links
```

```python
#определим содержимое 1 страницы
response = requests.get('https://lifehacker.ru/podderzhka-smartfonov-pixel/')
soup = BeautifulSoup(response.text, 'lxml')

head = soup.find('h1', class_='article-card__title').text
full_content = soup.find('section', class_='single-article__content').text.replace(soup.find('div', class_='read-also__container').text, '').replace(soup.find('div', class_='article-card single-article-card').text,'').replace('\n','').replace('\xa0',' ')
only_text = full_content.replace(soup.find('footer', class_='article-note article-note').text, '').replace("   Избранное       ",'')
```

```python
#получим для 10 страниц
result = []

for url in tqdm(links):
    article = {}
    article_url = f'{base_url}{url}' # подставляем часть ссылки, ведущую материал
    response = requests.get(article_url)
    soup = BeautifulSoup(response.text, 'lxml')

    head = soup.find('h1', class_='article-card__title').text
    full_content = soup.find('section', class_='single-article__content').text.replace(soup.find('div', class_='read-also__container').text, '').replace(soup.find('div', class_='article-card single-article-card').text,'').replace('\n','').replace('\xa0',' ')
    only_text = full_content.replace(soup.find('footer', class_='article-note article-note').text, '').replace("   Избранное       ",'')

    article['title'] = head
    article['text'] = only_text


    result.append(article)
```

```python
pd.set_option('display.max_colwidth', 400)
df = pd.DataFrame(result)
df
```
